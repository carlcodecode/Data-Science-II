```python
#Load IMDB Dataset

from datasets import load_dataset

dataset = load_dataset("imdb")

print(dataset)
```

```python
# What the data looks like

print(dataset["train"][67])
```

```python
from collections import defaultdict

# Postings list
postings = defaultdict(set)

# Iterate through the
for i, doc in enumerate(dataset["train"]):
  tokens = doc["text"].lower().split() # tokenize words in each doc
  for token in set(tokens):
    postings[token].add(i)

# Boolean Search
def boolean_search(aspect_terms, opinion_terms):
  # (aspect_1 OR aspect_2) AND (opinion_1 OR opinion_2)
  aspect_sets = [postings[t] for t in aspect_terms if t in postings]
  opnion_sets = [postings[t] for t in opinion_terms if t in postings]

  result_set = (set().union(*aspect_sets)).intersection(set().union(*opnion_sets))

  return result_set

for result in list(boolean_search(["story"], ["good"]))[:10]:
  print(dataset["train"]["text"][result])

```
